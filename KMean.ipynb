```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X,y = make_blobs(
    n_samples=100,
    n_features= 2,
    centers = 3, 
    cluster_std= 1, 
    shuffle= True, 
    random_state= None
)
```

```python
plt.scatter(
    X[:,0], X[:,1],
    c='white', marker='o',
    edgecolors='black', s=50
)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
model = kmeans.fit_predict(X)
```

```python
plt.scatter(
    X[model == 0, 0], X[model == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[model == 1, 0], X[model == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[model == 2, 0], X[model == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

plt.scatter(
    kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
# plot the centroids
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```
